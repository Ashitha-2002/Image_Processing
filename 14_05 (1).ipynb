```python
import cv2

img = cv2.imread("sunflower.jpg", 0)
print("Original matrix\n",img)

rows, cols = img.shape

sum1 = 0

for i in range(rows):
    for j in range(cols):
        sum1 += img[i,j]
mean = sum1 / (rows * cols)
print("Mean :: ", mean)

varianceSum = 0
for i in range(rows):
    for j in range(cols):
        varianceSum += (img[i,j] - mean) ** 2
variance = varianceSum / (rows * cols)
print("Variance :: ", variance)
```

```python
import cv2

A = [[2, 5, 3],
    [6, 1, 8],
    [4, 7, 8]]
print("Original matrix\n",A)

sum1 = 0

for i in range(3):
    for j in range(3):
        sum1 += A[i][j]
mean = sum1 / 9
print("Mean :: ", mean)

varianceSum = 0
for i in range(3):
    for j in range(3):
        varianceSum += (A[i][j] - mean) ** 2
variance = varianceSum / 9
print("Variance :: ", variance)
```

```python
import cv2
import os
import matplotlib.pyplot as plt
    

root_folder = "C:\\Users\\User\\Image_Processing\\Images"
imgList = []
for file in os.listdir(root_folder): 
    if file.endswith(('.jpg', '.jpeg', '.png', '.jfif')):
        file_path = os.path.join(root_folder,file)
        imgList.append(file_path)


means = []
variances = []

plt.figure(figsize=(15, 13))

for i in range(len(imgList)):
    print(imgList[i])
    imgOb = cv2.imread(imgList[i])
    rgb_image = cv2.cvtColor(imgOb, cv2.COLOR_BGR2RGB) 
    plt.subplot(5, 5, i+1)
    plt.imshow(rgb_image)
    
    gray_image = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2GRAY) 
    print(gray_image.shape)   
```

```python
def find_mean_variance(img):
    rows, cols = img.shape
    
    sum1 = 0

    for i in range(rows):
        for j in range(cols):
            sum1 += img[i,j]
    mean = sum1 / (rows * cols)
    
    varianceSum = 0
    
    for i in range(rows):
        for j in range(cols):
            varianceSum += (img[i,j] - mean) ** 2
    variance = varianceSum / (rows * cols)
    
    return mean, variance    
    
    
mean, variance = find_mean_variance(gray_image)
    
means.append(mean)
variances.append(variance)

print(means)
print(variances)
plt.scatter(means,variances)
```
