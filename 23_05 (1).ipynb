```python
import os
import cv2
import matplotlib.pyplot as plt

def find_mean_variance(img):
    rows, cols = img.shape
    
    sum1 = 0

    for i in range(rows):
        for j in range(cols):
            sum1 += img[i,j]
    mean = sum1 / (rows * cols)
    
    varianceSum = 0
    
    for i in range(rows):
        for j in range(cols):
            varianceSum += (img[i,j] - mean) ** 2
    variance = varianceSum / (rows * cols)
    
    return mean, variance 
    
root_folder = "C:\\Users\\User\\Image_Processing\\Images"
imgList = []
imgCount = 0
for file in os.listdir(root_folder): 
    if imgCount == 20:
        break
    if file.endswith(('.jpg', '.jpeg', '.png', '.jfif')):
        file_path = os.path.join(root_folder,file)
        imgList.append(file_path)
        imgCount += 1

means = []
variances = []

for i in range(len(imgList)):
    print(imgList[i])
    imgOb = cv2.imread(imgList[i]) 
    
    gray_image = cv2.cvtColor(imgOb, cv2.COLOR_BGR2GRAY) 
    mean, variance = find_mean_variance(gray_image)
    
    means.append(mean)
    variances.append(variance)

print("Means: ", means)
print("Variances : ", variances)
plt.scatter(means,variances)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load images
image1 = cv2.imread('bird.jpg')
image2 = cv2.imread('dog.jpg')
rows, cols = image2.shape[:2]

# Ensure the images are of the same size
image1 = cv2.resize(image1, (cols,rows))

added_image = cv2.add(image1, image2)
sub_image = cv2.subtract(image2, image1)
multiplied_image = cv2.multiply(image1, image2)

image1_float = image1.astype(np.float32)
image2_float = image2.astype(np.float32) + 1  # Adding 1 to avoid division by zero

divided_image = cv2.divide(image1_float, image2_float)

# Convert back to uint8
divided_image = cv2.normalize(divided_image, None, 0, 255, cv2.NORM_MINMAX)
divided_image = divided_image.astype(np.uint8)

plt.figure(figsize = (8,12))
plt.subplot(3,2,1)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(3,2,2)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(3,2,3)
plt.imshow(cv2.cvtColor(added_image, cv2.COLOR_BGR2RGB))
plt.title('Added Image')

plt.subplot(3,2,4)
plt.imshow(cv2.cvtColor(sub_image, cv2.COLOR_BGR2RGB))
plt.title('Subtracted Image')

plt.subplot(3,2,5)
plt.imshow(cv2.cvtColor(multiplied_image, cv2.COLOR_BGR2RGB))
plt.title('Multiplied Image')

plt.subplot(3,2,6)
plt.imshow(cv2.cvtColor(divided_image, cv2.COLOR_BGR2RGB))
plt.title('Divided Image')
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Function to threshold image to binary per channel
def threshold_image(image, threshold=128):
    # Split the image into its color channels
    channels = cv2.split(image)
    binary_channels = []
    for channel in channels:
        _, binary_channel = cv2.threshold(channel, threshold, 255, cv2.THRESH_BINARY)
        binary_channels.append(binary_channel)
    return cv2.merge(binary_channels)

# Load images
image1 = cv2.imread('cat.jpg')
image2 = cv2.imread('dog.jpg')

# Ensure the images are of the same size
image1 = cv2.resize(image1, (image2.shape[1], image2.shape[0]))

# Threshold images to create binary images
binary_image1 = threshold_image(image1)
binary_image2 = threshold_image(image2)

# Perform logical AND operation
logical_and = np.logical_and(binary_image1, binary_image2).astype(np.uint8) * 255

# Perform logical OR operation
logical_or = np.logical_or(binary_image1, binary_image2).astype(np.uint8) * 255

# Perform logical NOT operation on the first image
logical_not = np.logical_not(binary_image1).astype(np.uint8) * 255

# Display the results
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.title('Original Image 1')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Original Image 2')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(logical_and, cv2.COLOR_BGR2RGB))
plt.title('Logical AND')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(logical_or, cv2.COLOR_BGR2RGB))
plt.title('Logical OR')

plt.figure()
plt.imshow(cv2.cvtColor(logical_not, cv2.COLOR_BGR2RGB))
plt.title('Logical NOT')
plt.show()

```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Function to add two images manually
def add_images(image1, image2):
    # Ensure the images have the same dimensions
    if image1.shape != image2.shape:
        raise ValueError("Images must have the same dimensions for addition.")
    
    # Convert images to int32 to handle overflow
    image1_int = image1.astype(np.int32)
    image2_int = image2.astype(np.int32)
    
    # Create an empty array for the output image
    result_image = np.zeros_like(image1_int)
    
    rows, cols, channels = image1_int.shape
    
    # Manually add pixel values
    for row in range(rows):
        for col in range(cols):
            for channel in range(channels):
                pixel_sum = image1_int[row, col, channel] + image2_int[row, col, channel]
                result_image[row, col, channel] = np.clip(pixel_sum, 0, 255)
    
    # Convert back to uint8
    result_image = result_image.astype(np.uint8)
    
    return result_image

# Load images
image1 = cv2.imread('dog.jpg')
image2 = cv2.imread('bird.jpg')

# Ensure the images are of the same size
image1 = cv2.resize(image1, (image2.shape[1], image2.shape[0]))

# Add images manually
added_image = add_images(image1, image2)

# Display the result
plt.figure(figsize=(12, 8))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.title('Original Image 1')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Original Image 2')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(added_image, cv2.COLOR_BGR2RGB))
plt.title('Added Image')
```
